```python
## üéØ Ïã§Ïäµ Î™©Ìëú

- `.apply(lambda)` + `if-elif-else` Ï°∞Í±¥Î¨∏ Íµ¨Ï°∞ ÏùµÌûàÍ∏∞
- Ï§ëÏ≤© ÎîïÏÖîÎÑàÎ¶¨ÏóêÏÑú ÏïàÏ†ÑÌïòÍ≤å `.get().get()`ÏúºÎ°ú Í∞í Ï∂îÏ∂ú
- Í∞ÄÍ≤©Í≥º Î∞©Î¨∏ ÌöüÏàòÎ•º Í∏∞Ï§ÄÏúºÎ°ú Í≥†Í∞ù Î∂ÑÎ•ò Ïª¨Îüº ÏÉùÏÑ±
- `assert`Î°ú Í≤∞Í≥º Í≤ÄÏ¶ùÌïòÎäî Ïã§Î¨¥Ìòï Í≤ÄÏ¶ù Ìå®ÌÑ¥ Ï≤¥Îìù

---

## ‚úÖ Ïã§Ïäµ Ï°∞Í±¥

1. **price_category Ïª¨Îüº ÏÉùÏÑ±**
    - Í∏∞Ï§Ä: info["items"][0]["price"]
    - Ï°∞Í±¥:
        - `>= 10000` ‚Üí `"Í≥†Í∞Ä"`
        - `>= 5000` ‚Üí `"Ï§ëÍ∞Ä"`
        - `< 5000` ‚Üí `"Ï†ÄÍ∞Ä"`

2. **visit_grade Ïª¨Îüº ÏÉùÏÑ±**
    - Í∏∞Ï§Ä: info["customer"]["visit_count"]
    - Ï°∞Í±¥:
        - `>= 15` ‚Üí `"Îã®Í≥®"`
        - `>= 5` ‚Üí `"ÏùºÎ∞ò"`
        - `< 5` ‚Üí `"Ïã†Í∑ú"`
```

```python
## Ï†ïÎãµÎãµ

from pathlib import Path
import pandas as pd
import json

# ÌååÏùº Í≤ΩÎ°ú ÏßÄÏ†ï
data_dir = Path("../week04/data")
orders_path = data_dir / "claude_orders.csv"

assert data_dir.exists(), "ÌååÏùºÏù¥ Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏäµÎãàÎã§"
assert orders_path.exists(), "CSV ÌååÏùºÏù¥ Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏäµÎãàÎã§"

df = pd.read_csv(orders_path)
assert not df.empty, "Îç∞Ïù¥ÌÑ∞Í∞Ä ÎπÑÏñ¥ ÏûàÏäµÎãàÎã§."

# JSON Î¨∏ÏûêÏó¥ ÌååÏã±
df["info"] = df["order_info"].apply(json.loads)

# Í∞ÄÍ≤© Í∏∞Î∞ò Î∂ÑÎ•ò Ïª¨Îüº ÏÉùÏÑ±
df["price_category"] = df["info"].apply(
    lambda x: "Í≥†Í∞Ä" if x.get("items", [{}])[0].get("price", 0) >= 10000
    else "Ï§ëÍ∞Ä" if x.get("items", [{}])[0].get("price", 0) >= 5000
    else "Ï†ÄÍ∞Ä"
)

# Î∞©Î¨∏ Îì±Í∏â Ïª¨Îüº ÏÉùÏÑ±
df["visit_grade"] = df["info"].apply(
    lambda x: "Îã®Í≥®" if x.get("customer", {}).get("visit_count", 0) >= 15
    else "ÏùºÎ∞ò" if x.get("customer", {}).get("visit_count", 0) >= 5
    else "Ïã†Í∑ú"
)

# Í≤∞Í≥º Ï∂úÎ†•
print(df["price_category"])
print(df["visit_grade"])

# Í≤ÄÏ¶ù
assert df["price_category"].tolist() == ["Ï†ÄÍ∞Ä", "Ï†ÄÍ∞Ä", "Ï†ÄÍ∞Ä", "Ï†ÄÍ∞Ä", "Ï†ÄÍ∞Ä"], "‚ùå Í∞ÄÍ≤© Ïπ¥ÌÖåÍ≥†Î¶¨ Î∂ÑÎ•ò Ïò§Î•ò"
assert df["visit_grade"].tolist() == ["Îã®Í≥®", "ÏùºÎ∞ò", "Îã®Í≥®", "Ïã†Í∑ú", "ÏùºÎ∞ò"], "‚ùå Î∞©Î¨∏ Îì±Í∏â Î∂ÑÎ•ò Ïò§Î•ò"
```

```python
## ÏßÅÏ†ëÏã§ÏäµÏäµ

from pathlib import Path
import pandas as pd
import json

## ÌååÏùº Í≤ΩÎ°ú ÏßÄÏ†ï

data_dir = Path("../week04/data")
orders_path = data_dir / "claude_orders.csv"

assert data_dir.exists(), "ÌååÏùºÏù¥ Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏäµÎãàÎã§"

df = pd.read_csv(orders_path)

assert not df.empty, "Îç∞Ïù¥ÌÑ∞Í∞Ä ÎπÑÏñ¥ ÏûàÏäµÎãàÎã§."

df["info"] = df["order_info"].apply(json.loads)

df["price_category"] = df["info"].apply(lambda x: "Í≥†Í∞Ä" if x.get("itemes",[{}][0]).get("price",0) >= 10000 
                                        else "Ï§ëÍ∞Ä" if x.get("itemes",[{}][0]).get("price",0) >= 5000
                                        else "Ï†ÄÍ∞Ä")

df["visit_grade"] = df["info"].apply(lambda x: "Îã®Í≥®" if x.get("customer",{}).get("visit_count",0) >= 15
                                     else "ÏùºÎ∞ò" if x.get("customer",{}).get("visit_count",0) >= 5
                                     else "Ïã†Í∑ú")

print(df["price_category"])
print(df["visit_grade"])
```
