# 🧠 SQL 예습 Day 25 - JOIN 기초 (INNER vs LEFT)

---

## 📌 학습 목표
- `INNER JOIN`과 `LEFT JOIN`의 차이 이해
- 테이블 간 관계를 기반으로 실무형 조인 구조 학습

---

## 🧾 테이블 구조 예시

### `orders` 테이블

| order_id | customer_id | amount |
|----------|-------------|--------|
| 1        | 101         | 5000   |
| 2        | 102         | 3000   |
| 3        | 103         | 7000   |

---

### `customers` 테이블

| customer_id | name    |
|-------------|---------|
| 101         | Alice   |
| 102         | Bob     |
| 104         | Daniel  |

---

## 🧪 JOIN 실습 쿼리

### 🔹 INNER JOIN

```sql
SELECT 
  o.order_id,
  o.amount,
  c.name
FROM 
  orders o
INNER JOIN 
  customers c
ON 
  o.customer_id = c.customer_id;

  ✅ 결과: 양쪽 모두 매칭되는 행만 출력됨 (교집합)


### 🔹 LEFT JOIN

SELECT 
  o.order_id,
  o.amount,
  c.name
FROM 
  orders o
LEFT JOIN 
  customers c
ON 
  o.customer_id = c.customer_id;

✅ 결과: 모든 주문(order) 출력됨
➡️ 고객 정보 없는 경우 name = NULL로 표시됨


🧠 JOIN 정리표
JOIN 종류	기준 테이블	결과 설명
INNER JOIN	양쪽 모두	양쪽 모두 매칭되는 행만 출력 (교집합)
LEFT JOIN	왼쪽(orders)	왼쪽 전체 포함, 오른쪽은 NULL 허용
