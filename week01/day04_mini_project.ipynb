```python
grades = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}
scores = []
total = 0
# 학생 수를 입력받습니다.
student_count = int(input("총 학생 수를 입력하세요:"))

for i in range(student_count):
    # 학생의 점수를 입력받습니다.
    score = int(input(f"{i + 1}번째 학생의 점수를 입력하세요:"))
    print(f"{i + 1}번 학생 점수: {score}")
    
    if score >= 90:
        grades["A"] += 1
    elif score >= 80:
        grades["B"] += 1
    elif score >= 70:
        grades["C"] += 1
    elif score >= 60:
        grades["D"] += 1
    else:
        grades["F"] += 1
    
    total += score
    
    scores.append(score)
    
    
average = total / student_count

print(f"전체 평균 점수는 {average}점입니다.")
print(f"최고 점수는 {max(scores)}점입니다.")
print(f"학점 분포: {grades}")


```

```python
# ✅ 점수 입력 → 학점 분류 → 평균/최고점/분포 출력 (입력 유효성 검사 포함)

grades = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}
scores = []
total = 0

student_count = int(input("총 학생 수를 입력하세요: "))

for i in range(student_count):
    while True:
        try:
            score = int(input(f"{i + 1}번째 학생의 점수를 입력하세요 (0~100): "))
            
            # ✅ 유효성 검사: 0~100 사이의 점수인지 확인
            if 0 <= score <= 100:
                break
            else:
                print("⚠️ 점수는 0 이상 100 이하의 정수여야 합니다. 다시 입력해주세요.")
        except ValueError:
            print("❌ 숫자로만 입력해주세요. 다시 시도해주세요.")
    
    print(f"{i + 1}번 학생 점수: {score}")

    # ✅ 학점 분류
    if score >= 90:
        grades["A"] += 1
    elif score >= 80:
        grades["B"] += 1
    elif score >= 70:
        grades["C"] += 1
    elif score >= 60:
        grades["D"] += 1
    else:
        grades["F"] += 1

    total += score
    scores.append(score)

# ✅ 평균 점수 출력 (소수점 2자리)
average = total / student_count
print(f"\n📊 전체 평균 점수는 {average:.2f}점입니다.")

# ✅ 최고 점수 출력
print(f"🏆 최고 점수는 {max(scores)}점입니다.\n")

# ✅ 학점 분포 보기 좋게 출력
print("📈 학점 분포:")
for grade, count in grades.items():
    print(f"{grade}: {count}명")
```
